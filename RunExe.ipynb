```python
import subprocess
```

```python
def convert_data(cmd):
    with subprocess.Popen(cmd, stdout=subprocess.PIPE, stderr=subprocess.STDOUT) as p:
        for line in p.stdout.readlines():
            print(line.decode('utf-8'))
```

```python
root_path = '/home/salt/Documents/data/'
root_files = ['LOT0_PROC0_1.root', 'LOT1_PROC1_1.root', 'LOT2_PROC2_1.root', 'LOT3_PROC3_1.root']
converter = '/home/salt/Workspace/SemiTool/DataConverter/DataConverter/Debug/DataConverter'
target_path = './output/'
```

```python
for file in root_files:
    cmd = [converter, root_path + file, target_path]
    convert_data(cmd)
```
