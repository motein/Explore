```python
result = list(Path(mypath).rglob("*.xls*"))
```

```python
import pandas as pd
```

```python
cols = ['身份证号', '联系电话（手机）']
dd = pd.read_excel(str(result[4]), header=1)[cols]
for k, row in dd.iterrows():
    print(k)
    print(row['身份证号'])
```

```python
valid_filename = []
for item in result:
    doc = pd.read_excel(str(item), header=1)
    temp = doc.keys()
    if '联系电话（手机）' in temp:
        valid_filename.append(str(item))
    else:
        print(item)
```

```python
id_phone_dict = {}
for filename in valid_filename:
    doc = pd.read_excel(str(filename), header=1)[cols]
#     print(len(doc))
    for k, row in doc.iterrows():
        id_phone_dict[str(row['身份证号']).strip()] = str(row['联系电话（手机）'])
```

```python
df = pd.read_excel(final_filename, header=1)
```

```python
df['身份证号']
```

```python
for k, row in df.iterrows():
    id = str(df['身份证号'][k]).strip()
    if id in id_phone_dict:
        phone = id_phone_dict[id]
        df['电话号码'][k] = phone
    else:
        print(id)
```
