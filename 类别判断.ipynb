```python
def get_invalid_data(clss, data):
    if clss == 'A':
        temp = data[data['S'] > 0.1] # S
        temp = temp[temp['AspectRatio'] > 5] # AspectRatio
    elif clss == 'B':
        temp = data[data['S'] < 0.1] # S
        temp = temp[temp['AspectRatio'] < 5] # AspectRatio
    elif clss == 'C':
        temp = data[data['S'] < 0.1] # S
        temp = temp[temp['AspectRatio'] >= 5] # AspectRatio
    elif clss == 'D':
        temp = data[data['S'] < 0.1] # S
        temp = temp[temp['AspectRatio'] < 5] # AspectRatio
    elif clss == 'Ds':
        temp = data[data['S'] > 0.02] # S
        temp = temp[temp['AspectRatio'] < 5] # AspectRatio
    else:
        return None
```
