```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()

rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)
ax.scatter(x, y, c=colors, s=sizes, alpha=0.3,cmap='viridis')

vx = list(np.linspace(-3,3,8))
hx = list(np.linspace(-3,3,8))

ax.vlines(vx, -3, 3, lw=0.5, linestyles ="dotted", color='grey', alpha=0.5)
ax.hlines(hx, -3, 3, lw=0.5, linestyles ="dotted", color='grey', alpha=0.5)

plt.show()
```
